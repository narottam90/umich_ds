# Querying a Series

```python
sports = {'Archery' : 'Bhutan',
          'Golf' : 'Scotland',
          'Sumo' : 'Japan',
          'Taekwondo' : 'Korea'}

s = pd.Series(sports)
s
```

```python
s.iloc[3] # query by numeric location, starting at 0
```

```python
s.loc['Golf'] # query by the index label
```

```python
s[3]
```

```python
s['Golf']
```

```python
sports = { 99 : 'Bhutan',
           100 : 'Scotland',
           101 : 'Japan',
           102 : 'Korea'
           }

s = pd.Series(sports)
```

```python
s[0] # this doesn't call s.iloc[0] as expected, and generates a key error
```

## Iterating over a series

```python
s = pd.Series([100.00, 120.00, 101.00, 3.00])
s
```

```python
total = 0
for item in s:
    total += item
print(total)
```

```python
import numpy as np

total = np.sum(s)
print(total)
```

```python
# test speed between iterating and functional programming

s = pd.Series(np.random.randint(0, 1000, 10000))
s.head()
```

```python
len(s)
```

```python
%%timeit -n 100
summary = 0
for item in s:
    summary += item
```

```python
%%timeit -n 100
summary = np.sum(s)
```

```python
s += 2 # adds two to each item in s using broadcasting
s
```

```python
for label, value in s.iteritems():
    s.set_value(label, value+2) # set_value has been deprecated from pandas

s.head()
```

```python
%%timeit -n 10
for label, value in s.iteritems():
    s.loc[label] = value + 2
```

```python
%%timeit -n 10
s = pd.Series(np.random.randint(0, 1000, 10000))
s+=2
```
