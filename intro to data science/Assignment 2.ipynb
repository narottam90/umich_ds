```python
def answer_five():

    # Create a filter for the state level
    census_df_copy = census_df[census_df['SUMLEV'] == 50]

    # Create an index based on state names
    census_df_copy = census_df_copy.set_index(['STNAME'])

    # Count the number of unique counties using value_counts() and return the string at index 0 
    return census_df_copy.index.value_counts().index[0]

answer_five()
```

### Question 6
**Only looking at the three most populous counties for each state**, what are the three most populous states (in order of highest population to lowest population)? Use `CENSUS2010POP`.

*This function should return a list of string values.*

```python
def answer_six():
    lst = []
    census_df_copy = census_df[census_df['SUMLEV'] == 50]
    census_df_copy = census_df_copy.set_index(['STNAME'])

    state_df = pd.DataFrame()

    # get the unique state names
    state_df['State'] = census_df_copy.index.unique()

    # initialise all populations to 0
    state_df['Top3Pop'] = 0

    # set the index to the states
    state_df.set_index('State', inplace = True)

    # loop over all the states in state_df
    for state in state_df.index:
        county_pop = census_df_copy['CENSUS2010POP'][state]

        # check if county_pop has more tha one number, needed for states with one county only
        if (type(county_pop)) == pd.Series:
            stsum = sum(county_pop.sort_values(ascending = False)[:3])
        else:
            stsum = county_pop
        
        # set the 'Top3Pop' value to the sum of the top 3 most populous counties
        state_df['Top3Pop'].loc[state] = stsum

    for state in state_df.sort_values(by = 'Top3Pop', ascending = False).index[:3]:
        lst.append(state)
    
    return lst

answer_six()
```

### Question 7
Which county has had the largest absolute change in population within the period 2010-2015? (Hint: population values are stored in columns POPESTIMATE2010 through POPESTIMATE2015, you need to consider all six columns.)

e.g. If County Population in the 5 year period is 100, 120, 80, 105, 100, 130, then its largest change in the period would be |130-80| = 50.

*This function should return a single string value.*

```python
def answer_seven():
    county_df = pd.DataFrame()

    columns = ['CTYNAME', 'POPESTIMATE2010', 'POPESTIMATE2011', 'POPESTIMATE2012', 'POPESTIMATE2014', 'POPESTIMATE2015']

    # Create a county database with populatuion estimates, with index set to CTYNAME
    census_df_copy = census_df[census_df['SUMLEV'] == 50]
    county_df = census_df_copy[columns]
    county_df.set_index('CTYNAME', inplace=True)

    county_df['PopDiff'] = None

    for i in range(len(county_df.index)):
        max = county_df.iloc[i].max()
        min = county_df.iloc[i].min()
        county_df['PopDiff'].iloc[i] = max - min

    return county_df.sort_values(by = 'PopDiff', ascending=False).index[0]

answer_seven()
```

### Question 8
In this datafile, the United States is broken up into four regions using the "REGION" column. 

Create a query that finds the counties that belong to regions 1 or 2, whose name starts with 'Washington', and whose POPESTIMATE2015 was greater than their POPESTIMATE 2014.

*This function should return a 5x2 DataFrame with the columns = ['STNAME', 'CTYNAME'] and the same index ID as the census_df (sorted ascending by index).*

```python
def answer_eight():
    region_df = pd.DataFrame()

    census_df_copy = census_df[census_df['SUMLEV'] == 50]

    region_df = census_df_copy[   ((census_df_copy['REGION'] == 1) | (census_df_copy['REGION'] == 2)) & 
            (census_df_copy['POPESTIMATE2015'] > census_df_copy['POPESTIMATE2014']) &
            (census_df_copy['CTYNAME'].str.startswith('Washington'))]
        
    return region_df[['STNAME', 'CTYNAME']]

answer_eight()
```
